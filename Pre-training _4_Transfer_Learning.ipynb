SemEval 2015 had a task of training models for Aspect Based Sentiment Analysis. One of the datasets they provided is annotated restaurant reviews, with specific entities and their attributes.

Six entities were used in this dataset:
- **Food**;
    - Dishes or dining options;
- **Drinks**;
    - Specific drinks or drinking options;
- **Service**;
    - the (customer/kitchen/counter) service, on the promptness and quality of the restaurant’s service in general, the food preparation, the staff’s attitude and professionalism, the wait time, the options offered (e.g. takeout);
- **Ambience**;
    - the atmosphere or the environment of the restaurant’s interior or exterior space (e.g. terrace, yard, garden), the décor, entertainment options;
- **Location**;
    - the location of the reviewed restaurant in terms of its position, the surroundings, the view;
- **Restaurant** (I think it is more proper to call this business);
    - the restaurant as whole not focusing on any of the above five entity types.

Five attibutes are used in this dataset:
- **General**: (service, ambience, and location can only be assigned to this)
- **Prices**: prices of the food, the drinks or the restaurant in general
- **Quality**: the taste, the freshness, the texture, the consistency, the temperature, the preparation, the authenticity, the cooking or general quality of the food and the drinks served in the restaurant
- **Style&Options**: the presentation, the serving style, the portions size, the food/menu options or variety (e.g. innovative dishes/drinks, vegetarian options) of the food and of the drinks served in the restaurant
- **Miscellaneous**: attributes that do not fall into any of the aforementioned cases.

The dataset is provided as a standalone XML.
